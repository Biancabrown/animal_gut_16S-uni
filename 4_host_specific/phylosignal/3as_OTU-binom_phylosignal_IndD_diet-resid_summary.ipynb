# Load

```r
# whole dataset
physeq = readRDS(physeq_file)
physeq
```

```r
# metadata
metadata = physeq %>%
    sample_data %>%
    as.matrix %>%
    as.data.frame %>%
    dplyr::select(X.SampleID, scientific_name, class, order, family, genus, diet)

metadata$Sample_ID = rownames(metadata)
metadata %>% dfhead
```

```r
# lipa results
lipa_res = readRDS(lipa_res_file) %>%
    do.call(rbind, .)
lipa_res %>% dfhead
```

# Formatting

```r
# adding host metadata
lipa_res %>% nrow %>% print

lipa_res = lipa_res %>%
    inner_join(metadata, c('host'='X.SampleID')) 
               
lipa_res %>% dfhead
```

## Significant LIPA OTUs

```r
# q-value adjustment
lipa_res = lipa_res %>%
    group_by(subsample_rep) %>%
    mutate(pvalue = pvalue %>% as.Num,
           qvalue = p.adjust(pvalue, method='BH')) %>%
    ungroup() 

lipa_res$qvalue %>% summary %>% print
lipa_res %>% dfhead
```

```r
## all OTUs with sig p-value in >= 95% of reps
lipa_res = lipa_res %>%
    group_by(OTU, scientific_name) %>%
    summarize(median_coef = median(coef),
              mean_coef = mean(coef),
              sd_coef = sd(coef),
              qvalue = (length(qvalue) - sum(qvalue < 0.05)) / length(qvalue)) %>%
    ungroup() 

lipa_res$qvalue %>% summary %>% print
lipa_res %>% dfhead
```

```r
# filtering to just significant values
lipa_res = lipa_res %>%
    filter(qvalue < 0.05) 

n_sig_OTUs = lipa_res$OTU %>% unique %>% length
cat('Number of sig. OTUs:', n_sig_OTUs, '\n')
lipa_res %>% dfhead
```

```r
# OTUs per host species
lipa_res_s = lipa_res %>%
    group_by(scientific_name) %>%
    summarize(n = OTU %>% unique %>% length) %>%
    ungroup()

lipa_res_s$n %>% summary %>% print
lipa_res_s$scientific_name %>% length %>% print
```

# Table of OTU - host associations

```r
lipa_res_s = lipa_res %>%
    dplyr::select(OTU, scientific_name, mean_coef) %>%
    spread(OTU, mean_coef) %>%
    as.data.frame

rownames(lipa_res_s) = lipa_res_s$scientific_name
lipa_res_s$scientific_name = NULL

lipa_res_s %>% dfhead
```

```r
out_file = file.path(work_dir, 'OTU-binom_lipaMoran_IndD_sens_OTU-host-meanCoef.tsv')
write.table(lipa_res_s, file=out_file, sep='\t', quote=FALSE)
cat('File written:', out_file, '\n')
```

# Parsing blast results

**Length cutoff:** `Match alignment length is >=95 % length of query`

```r
# load table & filtering
blast_res = data.table::fread(blast_out_taxR_file, sep='\t') %>%
    filter(V4 / V7 >= hit_len_cutoff,
           V3 >= 90)

colnames(blast_res) = c(
    'qaccver',
    'saccver',
    'pident',
    'length',
    'mismatch',
    'gapopen',
    'qlen',
    'qstart',
    'qend',
    'slen',
    'sstart',
    'send',
    'evalue',
    'staxid',
    'lineage_original',
    'lineage')

blast_res %>% dfhead
```

```r
blast_res %>%
    filter(!grepl('unclassified phylum', lineage)) %>%
    .$lineage %>% unique %>% sort
```

```r
blast_res = blast_res %>%
    separate(lineage, c('Root_nt', 'Domain_nt', 'Phylum_nt', 'Class_nt', 
                        'Order_nt', 'Family_nt', 'Genus_nt', 'Species_nt'), sep=';')

blast_res %>% dfhead
```

### Filtering

```r
blast_res_f = blast_res %>%
    filter(!grepl('unclassified', Phylum_nt))

blast_res_f %>% dfhead
```

## Adding taxonomy

```r
tax = physeq_file %>%
    readRDS() %>%
    tax_table() %>%
    as.matrix %>%
    as.data.frame() %>%
    mutate(OTU = rownames(.)) %>%
    dplyr::select(-Species)

tax %>% dfhead
```

```r
blast_res_f = blast_res_f %>%
    inner_join(tax, c('qaccver'='OTU'))
tax = NULL

blast_res_f %>% dfhead
```

### Summary LIPA-taxonomy table

# 16S distance vs LIPA values

* x-axis: 16S distance among LIPA OTUs
* y-axis: LIPA values (mean +/- sd) 

## 16S distance

```r
lipa_OTUs = lipa_res$OTU %>% unique %>% gsub('^X([0-9])', '\\1', .)
cat('Number of LIPA OTUs: ', lipa_OTUs %>% length, '\n')
```

```r
# re-loading lipa results
tmp = gsub('^([0-9])', 'X\\1', lipa_OTUs)
lipa_res = readRDS(lipa_res_file) %>%
    do.call(rbind, .) %>%
    filter(OTU %in% tmp)

lipa_res$OTU %>% unique %>% length %>% print
lipa_res %>% dfhead
```

```r
# adding host metadata
lipa_res %>% nrow %>% print

lipa_res = lipa_res %>%
    inner_join(metadata, c('host'='X.SampleID')) 
               
lipa_res %>% dfhead
```

```r
## mean coef per subsample rep
lipa_res = lipa_res %>%
    group_by(OTU, scientific_name) %>%
    summarize(mean_coef = mean(coef)) %>%
    ungroup() 

lipa_res$OTU %>% unique %>% length %>% print
lipa_res %>% dfhead
```

```r
## pearson coef between each OTU
lipa_res_s = lipa_res %>%
    inner_join(lipa_res, c('scientific_name')) %>%
    group_by(OTU.x, OTU.y) %>%
    summarize(pearson_coef = cor(mean_coef.x, mean_coef.y, method='pearson')) %>%
    ungroup() 

lipa_res_s$pearson_coef %>% summary %>% print
lipa_res_s %>% dfhead
```

```r
# 16S branch length
lipa_OTU_16S = physeq %>%
    subset_taxa(rownames(tax_table(physeq)) %in% lipa_OTUs) %>%
    phy_tree %>% 
    cophenetic %>% 
    as.dist %>% 
    as.matrix 

lipa_OTU_16S[upper.tri(lipa_OTU_16S)] = NA

lipa_OTU_16S = lipa_OTU_16S %>%
    as.data.frame %>%
    mutate(OTU_X = rownames(.)) %>%
    gather(OTU_Y, brlen, -OTU_X) %>%
    filter(!is.na(brlen)) %>%
    mutate(OTU_X = gsub('^([0-9])', 'X\\1', OTU_X),
           OTU_Y = gsub('^([0-9])', 'X\\1', OTU_Y))

lipa_OTU_16S %>% dfhead
```

```r
# joining
lipa_OTU_16S = lipa_OTU_16S %>%
    inner_join(lipa_res_s, c('OTU_X'='OTU.x',
                             'OTU_Y'='OTU.y')) 

lipa_OTU_16S %>% dfhead
```

```r
# plotting
p = lipa_OTU_16S %>%
    ggplot(aes(brlen, pearson_coef)) +
    geom_point(alpha=0.5, shape='O') +
    labs(x='16S rRNA tree branch length',
         y='Pearson coef. of LIPA coef. values') +
    theme_bw()

options(repr.plot.width=5, repr.plot.height=3.5)
plot(p)
```

## Similar, with diff LIPA coef. values

```r
# which are very similar in branch length? 
max_brlen = max(lipa_OTU_16S$brlen)
lipa_OTU_16S_s = lipa_OTU_16S %>%
    mutate(brlen_norm = brlen / max_brlen) %>%
    filter(brlen < 0.1) %>%
    filter(pearson_coef < 0) %>%
    arrange(brlen_norm)
    
lipa_OTU_16S_s %>% dfhead
```

```r
# adding taxonomy 
tax = physeq %>%
    tax_table %>%
    as.matrix %>%
    as.data.frame %>%
    mutate(OTU = gsub('^([0-9])', 'X\\1', rownames(.))) %>%
    filter(OTU %in% lipa_res$OTU) %>%
    dplyr::select(-Domain, -Species) %>%
    unite(Taxonomy, Phylum, Class, Order, Family, Genus, sep=';')

lipa_OTU_16S_s_s = lipa_OTU_16S_s %>%
    inner_join(tax, c('OTU_X'='OTU')) %>%
    inner_join(tax, c('OTU_Y'='OTU')) %>% 
    dplyr::select(OTU_X, OTU_Y, brlen_norm, pearson_coef, 
                 Taxonomy.x, Taxonomy.y) %>%
    arrange(brlen_norm)

lipa_OTU_16S_s_s
```

```r
# number of OTUs in table?
c(lipa_OTU_16S_s_s$OTU_X, lipa_OTU_16S_s_s$OTU_Y) %>% unique %>% length
```

### Writing table

```r
out_file = file.path(work_dir, 'OTU-binom_lipaMoran_IndD_sens_LIPA-pearson.tsv')
write.table(lipa_OTU_16S_s_s, out_file, sep='\t', quote=FALSE, row.names=FALSE)
cat('File written:', out_file, '\n')
```

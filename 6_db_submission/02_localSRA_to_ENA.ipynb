````r
# physeq sample metadata
metadata = physeq_file %>%
    readRDS %>%
    sample_data %>%
    as.matrix %>%
    as.data.frame %>%
    mutate(Sample = rownames(.)) %>%
    mutate(Sample = gsub('^X([0-9])', '\\1', Sample),
           Sample = gsub('\\.', '_', Sample)) %>%
    select(Sample, collection_date, country, latitude, longitude, sample_type)

metadata %>% dfhead
````
